```python
import pandas as pd

# 
df = pd.read_csv(r"C:\Users\wngud\lab\previous\텍데분\project\hr\csv파일\pre_기독교.csv")
df2 = pd.read_csv(r"C:\Users\wngud\lab\previous\텍데분\project\hr\csv파일\pre_불교.csv")
df3 = pd.read_csv(r"C:\Users\wngud\lab\previous\텍데분\project\hr\csv파일\pre_천주교.csv")
concat_df = pd.concat([df,df2,df3])
```

```python
concat_df['split_content'] = concat_df['pre_content'].apply(lambda x: x.split()) # LDA 위해 split 작업
```

```python
from gensim import corpora

# 토큰화된 문서를 사용하여 corpora.Dictionary 생성
word_dict = corpora.Dictionary(concat_df['split_content'])
word_dict.filter_extremes() # keep_n=1000

# doc2bow 메서드에 토큰화된 문서를 전달하여 corpus 생성
corpus = [word_dict.doc2bow(tokens) for tokens in concat_df['split_content']]
```

```python
[(idx,word) for idx, word in word_dict.items()]
```

```python
import gensim
```

```python
N_TOPICS = 10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics = N_TOPICS, 
                                           id2word=word_dict)
```

```python
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

- 가까운 애들끼리는 좀 연관되어 있느 토픽 그리고 멀리 떨어져있는 토픽들은 관련없는 토픽  
- 빨간색은 토픽 내에서 중요하게 쓰인 단어 
- 파란색은 토픽을 상관없이 모든 지금 다큐먼트 모든 문서 안에서 전체적으로 많이 쓰인 단어
- 빨간색이 큰데 파란색도 크다 그러면 그토픽에서 그 단어ㅏ 중요한 단어라고 해석하기는 어려움 <-다른 토픽에서도 많이 등장한 단어이기 떄문
- 각 문서들이 토픽들의 어떤 조합으로 되어 있는 지는 보기 힘들끼 때문에 아래 함수를 사용하여

```python
for i, topic_list in enumerate(ldamodel[corpus]):
    print(f'{i}번째 문서의 topic 비중: {topic_list}')
    if i==11:
        break
```
